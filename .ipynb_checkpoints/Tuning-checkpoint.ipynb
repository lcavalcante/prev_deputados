## Train/Test Split

```python
from sklearn.model_selection import train_test_split


data = df.copy()
target = data.pop('votos')

X_train, X_test, y_train, y_test = train_test_split(data, target, shuffle=True,
                                                    test_size=0.5, random_state=49)
```

## Fit Model on Training Data

```python
from sklearn.linear_model import LinearRegression


lr_train = LinearRegression(fit_intercept=True)
lr_train.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

def calc_ISE(X_train, y_train, model):
    '''returns the in-sample R^2 and RMSE; assumes model already fit.'''
    predictions = model.predict(X_train)
    mse = mean_squared_error(y_train, predictions)
    rmse = np.sqrt(mse)
    return model.score(X_train, y_train), rmse


    
def calc_OSE(X_test, y_test, model):
    '''returns the out-of-sample R^2 and RMSE; assumes model already fit.'''
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    return model.score(X_test, y_test), rmse
```

# Model Tuning

## Train/Validation/Test Split

```python
# intermediate/test split (gives us test set)
X_intermediate, X_test, y_intermediate, y_test = train_test_split(data, 
                                                                  target, 
                                                                  shuffle=True,
                                                                  test_size=0.2, 
                                                                  random_state=15)

# train/validation split (gives us train and validation sets)
X_train, X_validation, y_train, y_validation = train_test_split(X_intermediate,
                                                                y_intermediate,
                                                                shuffle=False,
                                                                test_size=0.25,
                                                                random_state=2018)
```

```python
# delete intermediate variables
del X_intermediate, y_intermediate

# print proportions
print('train: {}% | validation: {}% | test {}%'.format(round(len(y_train)/len(target),2),
                                                       round(len(y_validation)/len(target),2),
                                                       round(len(y_test)/len(target),2)))
```

```python
from sklearn.linear_model import Ridge, RidgeCV, LassoCV, LassoLarsCV
from sklearn.model_selection import cross_val_score



def calc_train_error(X_train, y_train, model):
    '''returns in-sample error for already fit model.'''
    predictions = model.predict(X_train)
    mse = mean_squared_error(y_train, predictions)
    rmse = np.sqrt(mse)
    return rmse
    
def calc_validation_error(X_test, y_test, model):
    '''returns out-of-sample error for already fit model.'''
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    return rmse
    
def calc_metrics(X_train, y_train, X_test, y_test, model):
    '''fits model and returns the RMSE for in-sample error and out-of-sample error'''
    model.fit(X_train, y_train)
    train_error = calc_train_error(X_train, y_train, model)
    validation_error = calc_validation_error(X_test, y_test, model)
    return train_error, validation_error
```

```python
lr = LinearRegression(fit_intercept=True)

train_error, test_error = calc_metrics(X_train, y_train, X_test, y_test, lr)
train_error, test_error = round(train_error, 3), round(test_error, 3)

print('train error: {} | test error: {}'.format(train_error, test_error))
print('train/test: {}'.format(round(train_error/test_error, 2)))
```

```python
alphas = [0,0.001, 0.01, 0.05, 0.1, 0.3, 1, 3, 5, 10, 15, 30, 50, 75]
print('All errors are RMSE')
print('-'*76)

new_train_error = []
new_validation_error = []
new_test_error = []


i = 0
for alpha in alphas:
    # instantiate and fit model
    ridge = Ridge(alpha=alpha, fit_intercept=True, random_state=99)
    ridge.fit(X_train, y_train)
    # calculate errors
    new_train_error.append(np.sqrt(mean_squared_error(y_train, ridge.predict(X_train))))
    new_validation_error.append(np.sqrt(mean_squared_error(y_validation, ridge.predict(X_validation))))
    new_test_error.append(np.sqrt(mean_squared_error(y_test, ridge.predict(X_test))))
    # print errors as report
    print('alpha: {:7} | train error: {:5} | val error: {:6} | test error: {}'.
          format(alpha,
                 round(new_train_error[i],3),
                 round(new_validation_error[i],3),
                 round(new_test_error[i],3)))
    i += 1
```

```python
fig, ax = plt.subplots()

ax.plot(alphas, new_train_error)
ax.plot(alphas, new_validation_error)
ax.plot(alphas, new_test_error)

ax.set(xlabel='lambda', ylabel='train_error')
ax.grid()
```

```python
# train/test split
X_train, X_test, y_train, y_test = train_test_split(data, 
                                                    target, 
                                                    shuffle=True,
                                                    test_size=0.2, 
                                                    random_state=15)

# instantiate model
ridge = Ridge(alpha=5, fit_intercept=True, random_state=99)

# fit and calculate errors
new_train_error, new_test_error = calc_metrics(X_train, y_train, X_test, y_test, ridge)
new_train_error, new_test_error = round(new_train_error, 3), round(new_test_error, 3)
```

```python
print('ORIGINAL ERROR')
print('-' * 40)
print('train error: {} | test error: {}\n'.format(train_error, test_error))
print('train/test: {}\n'.format(round(train_error/test_error ,2)))
print('ERROR w/REGULARIZATION')
print('-' * 40)
print('train error: {} | test error: {}'.format(new_train_error, new_test_error))
print('new_train/new_test: {}\n'.format(round(new_train_error/new_test_error ,2)))
```

## Cross Validation

```python
def rmse_cv(model):
    rmse= np.sqrt(-cross_val_score(model, X_train, y_train, scoring="neg_mean_squared_error", cv = 10))
    return(rmse)
```

```python
alphas = [0, 0.01, 0.05, 0.1, 0.3, 1, 3, 5, 10, 15, 30, 50, 75]

val_errors = []


val_errors = [rmse_cv(Ridge(alpha = alpha, fit_intercept=True, random_state=99)).mean() 
            for alpha in alphas]
```

```python
cv_ridge = pd.Series(val_errors, index = alphas)
```

```python
cv_ridge = pd.Series(cv_ridge, index = alphas)
cv_ridge.plot(title = "Validation")
plt.xlabel("alpha")
plt.ylabel("rmse")
```

```python
print('best alpha: {}'.format(alphas[np.argmin(val_errors)]))
```

```python
cv_ridge.min()
```

```python
from sklearn import neighbors



# intermediate/test split (gives us test set)
X_intermediate, X_test, y_intermediate, y_test = train_test_split(data, 
                                                                  target, 
                                                                  shuffle=True,
                                                                  test_size=0.2, 
                                                                  random_state=15)

# train/validation split (gives us train and validation sets)
X_train, X_validation, y_train, y_validation = train_test_split(X_intermediate,
                                                                y_intermediate,
                                                                shuffle=False,
                                                                test_size=0.25,
                                                                random_state=2018)

rmse_val = [] #to store rmse values for different k
for K in range(100):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(X_train, y_train)  #fit the model
    pred=model.predict(X_validation) #make prediction on test set
    error = np.sqrt(mean_squared_error(y_validation,pred)) #calculate rmse
    rmse_val.append(error) #store rmse values
    print('RMSE value for k= ' , K , 'is:', error)
```

```python
#plotting the rmse values against k values
curve = pd.DataFrame(rmse_val) #elbow curve 
curve.plot()
```

```python
knn_model = neighbors.KNeighborsRegressor(n_neighbors = 34)

knn_model.fit(X_train, y_train)



pred=knn_model.predict(X_train) #make prediction on test set
error_train = np.sqrt(mean_squared_error(y_train,pred)) #calculate rmse

pred=knn_model.predict(X_validation) #make prediction on test set
error_validation = np.sqrt(mean_squared_error(y_validation,pred)) #calculate rmse


pred=knn_model.predict(X_test) #make prediction on test set
error_test = np.sqrt(mean_squared_error(y_test,pred)) #calculate rmse

print('TRAIN ERROR')
print('-' * 40)
print(error_train)
print()
print('Validation ERROR')
print('-' * 40)
print(error_validation)
print()
print('TEST ERROR')
print('-' * 40)
print(error_test)
```
